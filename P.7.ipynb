```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A, B, C, D = 5, 1, 0, 0
n_points = 100
x = np.linspace(0, 6, n_points)
y = A * np.sin(B * x + C) + D # sine function
noise = np.random.normal(0, 2, n_points)
y_noise = y + noise
```

```python
plt.scatter(x, y_noise, label = "Noisy Data")
plt.plot(x, y, color = "red", label = "True Sine Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Generated Sine Dataset with Noise")
plt.show() 
```
