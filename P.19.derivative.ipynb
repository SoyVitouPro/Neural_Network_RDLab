```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 2 * x
```

```python
x = np.array( range ( 5 ))
y = f(x)
```

```python
plt.plot(x, y)
plt.legend()
plt.show()
```

- Derivative on f(x) = 2x -> f'(x) = 2

To fine slope this function is pretty easy, you can use this formula



```python
print ((y[ 1 ] - y[ 0 ]) / (x[ 1 ] - x[ 0 ]))
```

#### Now let's try with f(x) = 2x^2

```python
def f(x):
    return 2*x**2
```

- Repeat the same process

```python
x = np.array( range ( 5 ))
y = f(x)

print(x)
print(y)

plt.plot(x, y)
plt.show()

# x1 = 0, x2 = 1
print ((y[ 1 ] - y[ 0 ]) / (x[ 1 ] - x[ 0 ]))

# x1 = 1, x2 = 2
print ((y[ 2 ] - y[ 1 ]) / (x[ 2 ] - x[ 1 ]))
```

```python
x = np.arange(0, 5, 0.001)
y = f(x)

print(x)
print(y)

plt.plot(x, y)
plt.show()

# x1 = 0, x2 = 1
print ((y[ 1 ] - y[ 0 ]) / (x[ 1 ] - x[ 0 ]))

# x1 = 1, x2 = 2
print ((y[ 2 ] - y[ 1 ]) / (x[ 2 ] - x[ 1 ]))
```
