```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# linear function y = mx + b
m, b = 2, 1
n_points = 100

# random number from 0 to 10, and 100 points
x = np.linspace(0, 10, n_points)
y = m * x + b

# add noise random to y
noise = np.random.normal(0, 1, n_points)

y_noise = y + noise
```

```python
# ploting data
plt.scatter(x, y_noise, label='Noisy Data')
plt.plot(x, y, color="red", label = "True Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Generated Linear Dataset with Noise")
plt.show()
```
